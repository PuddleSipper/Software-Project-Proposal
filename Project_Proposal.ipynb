# Project Proposal

What problem will your game/simulation address?: My game/simulation will address issues regarding the decline in videogame creativity with the rise of AI and will be an adventure platformer

### Requirement outline
 - Inputs: Player inputs (WASD/arrow keys as movement)
 - Processing: Process player inputs
 - Outputs: Translate player inputs into movement/actions
 - Transmission: Data will not need to be transmitted
 - Storage: Save data will need to be stored

### Functional and Non-Functional requirements
Functional:
 - Inputs and movement will need to feel fluid and mistakes should feel like the player's own
 - Obstacles need to feel challenging but possible, player's should feel rewarded for overcoming them
 - Challenges need to be fun to do, and should reward the player when they do something correctly
Non-Functional:
 - The game should have little to no inconsistencies in its aesthetics
 - The game should be visually captivating, players shouldn't have to look at horribly made sprites or similar things

### Pseudocode and Flowcharts
Pseudocode:
 Normal Levels
 -User inputs movement <br>
 -User attempts to beat obstacles <br>
 -If user fails obstacles, restart from beginning of level or at savepoint and repeat from line 2 <br>
 -Elif user fails obstacles, restart from beginning of level or at savepoint, provide assistance with obstacles if it is the tutorial and repeat from line 2 <br>
 -Else user passes obstacles, reward them with new ability and some currency and move to next level <br>
 -If user has all abilities, reward with just currency and move to next level
 
 Boss Levels
 -User attempts to fight boss (through unlocked abilities) <br>
 -If user fails, restart from nearest savepoint and repeat from line 1 <br>
 -Else user succeeds, reward with methods to upgrade, such as more hit points or more damage, reward with currency as well
